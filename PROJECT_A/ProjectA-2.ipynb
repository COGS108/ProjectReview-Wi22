### Forbes' NFL Valuations
This data had to be scraped as it was not available as a CSV file, so the AJAX data was parsed and formatted into JSON to be read by pandas' read_json() function. Fortunately, there were only 32 observations per table and both tables were formatted the exact same way, so they could both be handled in the same loop.

```python
# specify URLs to create tables from Forbes' NFL valuations
articles = ['https://web.archive.org/web/20171130034303/https://www.forbes.com/ajax/list/data?year=2015&uri=nfl-valuations&type=organization',
            'https://web.archive.org/web/20171130034303/https://www.forbes.com/ajax/list/data?year=2016&uri=nfl-valuations&type=organization']
json_tables = []

# parse and create datasets for both 2015 and 2016
for article in articles: 
    req = requests.get(article)

    json_table = json.loads('{}') # create json file to store parsed entries
    dat = req.json() # store as json for iteration

    for ind, i in enumerate(dat): # iterate through each observation
        json_table.update({str(ind):i}) # format and store into json_table
        
    json_tables.append(json.dumps(json_table)) # store in list in json format

# create dataframes for both years
valuations_2015 = pd.read_json(json_tables[0], orient='index')
valuations_2016 = pd.read_json(json_tables[1], orient='index')
```

Additional formatting and cleaning is done to remove columns that are definitely not needed for any analysis relevant to the project, to rename columns for readability, and to format numerical values (money values were recorded in millions, which were not changed to keep some readability, but percentages were converted to decimals).

```python
# additional cleaning/formatting of the valuation dataframes
def formatdf(val_df):
    # print(val_df.dtypes)
    val_df = val_df.sort_values(by='rank')
    val_df = val_df.reset_index()
    val_df = val_df.drop(['index', 'position', 'rank', 'uri', 'imageUri', 'owners'], axis=1) # drop unnecessary columns
    val_df[['debtValue', 'oneYearValueChange']] = val_df[['debtValue', 'oneYearValueChange']].apply(lambda x: x * .01) # percentages to decimals
    val_df = val_df.rename(columns={"operatingIncome": "operating_income", "debtValue": "debt_over_value", "oneYearValueChange": "value_change", "valueList": "value"}) # rename columns
    return val_df
    
valuations_2015 = formatdf(valuations_2015)
valuations_2016 = formatdf(valuations_2016)
```

Next, two standardizing functions were created along with an additional column to track the total amount of Super Bowl wins a Super Bowl winner has. These will ensure that data such as team names or dates is in a consistent and appropriate format in order to make data analysis easier.

```python
### Create functions for standardizing/manipulating data ###

def standardize_date(string):
    # change months to number format
    string = string.replace('Jan','1')
    string = string.replace('Feb','2')
    # add backslashes
    string = string.replace(' ','/')
    
    return string

# Normalize team names that have changed (updated to current name)
def standardize_team_name(string):
    string = string.replace("Phoenix Cardinals", "Arizona Cardinals")
    string = string.replace("Washington Redskins", "Washington Football Team")
    string = string.replace("St. Louis Rams", "Los Angeles Rams")
    string = string.replace("Baltimore Colts", "Indianapolis Colts")
    string = string.replace("San Diego Chargers", "Los Angeles Chargers")
    string = string.replace("Los Angeles Raiders", "Las Vegas Raiders")
    string = string.replace("Oakland Raiders", "Las Vegas Raiders")
    string = string.replace("Houston Oilers", "Tennessee Titans")
    string = string.replace("Tennessee Oilers", "Tennessee Titans")

    return string

def cut_date(string):
    # remove dates so that only year remains
    string = int(string.split("/")[2])

    return string

def yr_2021_to_2020(string):
    string = string.replace("2021", "2020")

    return string

def get_city(city):
    city = city[0]
    
    return city
```

```python
#Standardize dates and normalize team names
superbowl['Date'] = superbowl['Date'].apply(standardize_date)
superbowl['Winner'] = superbowl['Winner'].apply(standardize_team_name)
superbowl['Loser'] = superbowl['Loser'].apply(standardize_team_name)

#Remove month/day from Data column
superbowl["Date"] = superbowl["Date"].apply(cut_date)
superbowl.sort_values(by = "Date", inplace = True)
superbowl.reset_index(drop = True, inplace = True)
```

```python
gamescores["team_home"] = gamescores["team_home"].apply(standardize_team_name)
gamescores["team_away"] = gamescores["team_away"].apply(standardize_team_name)

# 2021 years were changed to 2020 because they're part of the 2020 season
gamescores["schedule_date"] = gamescores["schedule_date"].apply(yr_2021_to_2020)
```

In order to track the success of each team in the NFL over the period of 1990-2020, the win rates of each team were calculated and tracked on a per-game basis. The average win rate per year of each team was tracked in a new dataset.

```python
# Ties are handled as a loss for both teams for simplicity's sake in the win rate calculation
# avg win rate was taken for each year

gamescores["win"] = np.nan
gamescores["win_rate_home"] = np.nan
gamescores["win_rate_away"] = np.nan
total_wins = {}
total_games = {}
win_rates = {}

# get win and win rate statistics for each game
for index, row in gamescores.iterrows():
    # initialize variables
    year = int(row["schedule_date"].split("/")[2])
    
    if(row["team_home"] not in total_wins):
        total_wins[row["team_home"]] = 0
    if(row["team_away"] not in total_wins):
        total_wins[row["team_away"]] = 0
    
    if(row["team_home"] not in total_games):
        total_games[row["team_home"]] = 0
        for yr in range(1990, 2021):
            win_rates[row["team_home"]] = {yr : []}
    if(row["team_away"] not in total_games):
        total_games[row["team_away"]] = 0
        for yr in range(1990, 2021):
            win_rates[row["team_away"]] = {yr : []}
    
    # calculate win and track win/loss
    if(row["score_home"] > row["score_away"]):
        gamescores.at[index, "win"] = 1
            
        total_wins[row["team_home"]] += 1
    
    elif(row["score_home"] < row["score_away"]):
        gamescores.at[index, "win"] = -1
        
        total_wins[row["team_away"]] += 1
        
    else:
        gamescores.at[index, "win"] = 0
    
    total_games[row["team_home"]] += 1
    total_games[row["team_away"]] += 1
    
    if year not in win_rates[row["team_home"]]:
        win_rates[row["team_home"]][year] = []
    if year not in win_rates[row["team_away"]]:
        win_rates[row["team_away"]][year] = []

    # update running win rates
    gamescores.at[index, "win_rate_home"] = total_wins[row["team_home"]] / total_games[row["team_home"]]
    gamescores.at[index, "win_rate_away"] = total_wins[row["team_away"]] / total_games[row["team_away"]]
    
    # store win rates with year
    win_rates[row["team_home"]][year].append(total_wins[row["team_home"]] / total_games[row["team_home"]])
    win_rates[row["team_away"]][year].append(total_wins[row["team_away"]] / total_games[row["team_away"]])
```

```python
# find average win rate for each team in each year
for team in win_rates:
    for year in win_rates[team]:
        try:
            win_rates[team][year] = statistics.mean(win_rates[team][year])
        except:
            win_rates[team][year] = 0
```

```python
# create dataframe with win_rate data
name_list = []
year_list = []
wr_list = []

for team in win_rates:
    for pair in win_rates[team]:
        name_list.append(team)
        year_list.append(pair)
        wr_list.append(win_rates[team][pair])

wr_df = pd.DataFrame(list(zip(name_list, year_list, wr_list)), columns = ["name", "year", "win_rate"])
wr_df.sort_values(by = "year", inplace = True)
wr_df.reset_index(drop = True, inplace = True)
```

### Win Rate and Unemployment Rate

One way that we can check for a relationship between team success and city economic status is by comparing a team's win rate to its city's unemployment rate. In order to determine if there is a relationship between win rate and unemployment, we need to merge the data into a dataframe that has what we need.

```python
# combine unemployment and win_rate data
wr_unemp = economy.merge(city_income, on = "county")
wr_df["city"] = wr_df["name"].map(nfl_teams)
wr_df["city"] = wr_df["city"].apply(get_city)
wr_unemp = wr_unemp.merge(wr_df, on = "city")

# drop unnecessary values
wr_unemp = wr_unemp[wr_unemp["year_x"] == wr_unemp["year_y"]]
wr_unemp.drop(["income mean", "state_ab", "mean", "median", "stdev", "sumw", "year_y", "zip"], axis = 1, inplace = True)

# rearrange and tidy dataframe
wr_unemp.drop_duplicates(subset = ["year_x", "city"], inplace = True)
wr_unemp = wr_unemp[["city", "county", "state", "name", "year_x", "win_rate", "average unemployment"]]
wr_unemp.rename({"average unemployment" : "avg_unemp", "year_x" : "year"}, axis = 1, inplace = True)
wr_unemp.reset_index(drop = True, inplace = True)
```

```python
wr_unemp
```

With the proper data all in the same place, we should visualize the data to get a sense of any potential relationships. To do this, we'll use a scatter matrix containing win rates and unemployment rates, separated by year.

```python
sns.pairplot(wr_unemp, hue = 'year', vars = ["win_rate", "avg_unemp"], palette = 'Paired', height = 4)
```

Win rate, as shown earlier, takes a normal distribution and is very similar for each year. The distribution of unemployment, however, varies greatly between years. When looking at the scatter plots of unemployment vs win rate, there doesn't appear to be a strong correlation aside from some outliers in each year. Based off the scatter matrix, it seems that for most cases unemployment rate is not affected by win rate and in general win rate is not a predictor of unemployment rate.

In order to further analyze whether or not a statistically significant relationship exists between win rate and unemployment, we'll perform linear regression on the win rates and unemployment rates for each year from 2011 to 2015.

```python
# Calculate and print regressions for 2011-2015
for yr in range(2011, 2016):
    print("Year:", yr)
    unemp_wr_reg_df = wr_unemp[["year", "win_rate", "avg_unemp"]]
    unemp_wr_reg_df = unemp_wr_reg_df[unemp_wr_reg_df["year"] == yr]
    outcome, predictors = patsy.dmatrices("win_rate ~ avg_unemp", unemp_wr_reg_df)

    mod_log = sm.OLS(outcome, predictors)

    fit_res = mod_log.fit()

    print(fit_res.summary())
    print("\n")
```

Based off the p-values returned by the linear regression for each year, we can determine that win rate is most likely not a statistically significant predictor of unemployment rate. While the p-value of the regression for the year 2012 is very slightly in the range of we would consider statistically significant (p < 0.05), the p-values of the other years are not so it is safe to assume that these results are not statistically significant. 

Overall, these results do not indicate that there is a relationship between team win rate and city unemployment rate.

**Limitations**

One of the biggest limitations of the study comes from the ambiguous nature of the qualities we tried to quantify. "Team success" and "city economic status" both are not specific quantities, but the holistic evaluation of many factors, of which we can only account for a few. We defined "team success" and "city economic status" as a team's value/revenue/wins and their city's unemployment rate/average household income respectively, but that is far from the complete picture. With few factors to work with, we can only answer the question on our own terms and our analysis cannot cover all the other factors that may define and affect those qualities.

Many limitations also arise from merging team stats datasets with the city's unemployment and income datasets. Having to merge multiple datasets together creates many restrictions on which data can be used across all the sets. The years have to line up, missing values may arise when merging datasets together, and other complications occur when dealing with vastly different datasets collected from different sources. As such, the subset of data that can actually be analyzed narrows down extremely, and most of the analysis can only be done from the years 2011-2015 (and especially 2014-2015). This severely impacts analysis because the less data is available, the less confidence any result or analysis can have. Despite both NFL team histories and city economics preferring lengthier periods of analysis for effects and trends over time to really be shown, the analysis has to deal with a very strict timeframe.

The restrictive timeframe and the datasets used also limits confidence because logically, changes in a city's economic status are gradual, and do not happen immediately. Revenue generated from a team's presence first reaches the city, and then it may trickle into creating new jobs or increasing household income (if at all). A more appropriate metric to analyze would be each city's yearly income (not household income), but it was difficult to find appropriate data, which limited us in finding such a set and we had to settle for less direct values.

The presence of confounders affecting the independent (team success) and dependent (city economic status) variables also greatly affects the analysis. Many outside factors affect a city's unemployment rate and average household income, which is reflected in their distributions relative to the team's revenue. Even if a team's success affects their city's unemployment rate/household income, it may be one factor among many others, and the impact it makes may be too small, relatively, to be noticeable statistically.

The limited timeframe also made it difficult to eliminate some confounders by analyzing one specific team for an extended amount of years, which might have lead to more productive models by dealing with trends in one team and one city over time, rather than analyzing the league as a whole.
